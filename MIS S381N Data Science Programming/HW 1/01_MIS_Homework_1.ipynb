## 1 Rock, Paper, Scissors
---
The popular game, you against the computer. Who will win?

```python
# Import module
import random
```

```python
# Function to get the user choice in each turn
def get_user_choice(round):
    """Ask the user for their choice in the Rock, Paper, Scissors game.
       Save the user choice.
       
       round : it is the number of round played so far in the Rock, Paper, Scissors game.
    """
    user_choice = input('\u001b[1mRound {}:\u001b[0m\nWould you like to Quit?...Q \nYour turn (R/P/S):'.format(round))
    user_choice = user_choice.upper()
    if user_choice not in ('R','P','S','Q'):
        print('\u001b[1mEnter a valid entry!\u001b[0m')
        user_choice = get_user_choice(round)
    return user_choice
```

```python
# Function to get the choice of the computer in each turn
def get_computer_choice(num_user_R,num_user_P,num_user_S,n_games):
    """Choose the best game for the computer based on randomness or user bias.
       - first round, computer choice is random.
       - If user is playing one option more often then computer choose to play the best option.
       - If user is playing two option more often and one less often, computer makes a random choice between 2 best options.
       
       num_user_R :  number of games that user played Rock
       num_user_P :  number of games that user played Paper
       num_user_S :  number of games that user played Scissors
       n_games :     number of games played so far
    """
    # Getting the maximum of choices played
    often_user_choice = max(num_user_R,num_user_P,num_user_S)
    
    # Random choice for first round or equal distribution of user choices
    if n_games <= 1 or (num_user_R == num_user_P and num_user_P == num_user_S):
        computer_choice = random.choice(['R','P','S'])
    else:
        # Two more often and equal
        if   num_user_R == num_user_P and (num_user_R+num_user_P)/n_games > 2/3: # Tie between R and P (both more often)
            computer_choice = 'P'
        elif num_user_R == num_user_S and (num_user_R+num_user_S)/n_games > 2/3: # Tie between R and S (both more often)
            computer_choice = 'R'
        elif num_user_P == num_user_S and (num_user_P+num_user_S)/n_games > 2/3: # Tie between P and S (both more often)
            computer_choice = 'S'
        # Only one more often
        elif num_user_R == often_user_choice: # More often R
            computer_choice = 'P'
        elif num_user_P == often_user_choice: # More often P
            computer_choice = 'S'
        elif num_user_S == often_user_choice: # More often S
            computer_choice = 'R'
        
    return computer_choice
```

```python
# Function to define who wins between the computer and the user in each turn
def who_wins(computer_choice,user_choice,user_won_games):
    """Evaluate who wins the current round in the Rock, Paper, and Scissors game.
       If user wins, add 1 to the number of won games by the user so far.
       
       computer_choice : it is the game of the computer in the round.
       user_choice :     it is the game of the user in the round.
       user_won_games :  number of won games by the user so far
    """
    if user_choice == 'Q':
        return -1, -1, -1
    else:
        if computer_choice == user_choice:
            message = 'Ohhh, '
            msg_result = 'It is a draw!'
        if computer_choice == 'R' and user_choice == 'S':
            message = 'Rock smashes Scissors, '
            msg_result = 'Computer wins!'
        if computer_choice == 'R' and user_choice == 'P':
            message = 'Paper covers Rock, '
            msg_result = 'You win!'
            user_won_games += 1
        if computer_choice == 'S' and user_choice == 'P':
            message = 'Scissors can cut up Paper, '
            msg_result = 'Computer wins!'
        if computer_choice == 'S' and user_choice == 'R':
            message = 'Rock smashes Scissors, '
            msg_result = 'You win!'
            user_won_games += 1
        if computer_choice == 'P' and user_choice == 'S':
            message = 'Scissors can cut up Paper, '
            msg_result = 'You win!'
            user_won_games += 1
        if computer_choice == 'P' and user_choice == 'R':
            message = 'Paper covers Rock, '
            msg_result = 'Computer wins!'
        return message, msg_result, user_won_games
```

```python
# Rock. Paper, Scissors game - function which iterate the game until the user quits.
def rock_paper_scissors():
    """This is the popular game Rock, Paper, an Scissors,
       we play in turns, computer play first and then it is your turn.
       Both, computer and user can choice one of the options: Rock (R), Paper (P), or Scissors (S)
       - Rock smashes Scissors, so Rock beats Scissors.
       - Scissors can cut up paper, so Scissors beat Paper.
       - Paper covers Rock, so Paper beats Rock.
       - If both players chose the same, it is a draw.
       If you enter a non valid option, then you have to re-enter a new valid one.
       You can quit the game with 'Q'.
    """
    # Set up inicial
    q = 0 # Indicator of quit (1:quit)
    n_games = 0 # Number of games
    num_user_R = 0 # Number of games in which user's played Rock
    num_user_P = 0 # Number of games in which user's played Paper
    num_user_S = 0 # Number of games in which user's played Scissors
    n_games_user_win = 0 # Number of games in which the user's won

    # Game start
    while q == 0:
        n_games += 1
        computer_choice = get_computer_choice(num_user_R,num_user_P,num_user_S,n_games)
        user_choice = get_user_choice(n_games)

        if user_choice == 'Q':
            q += 1
            n_games -= 1
            print('\u001b[1mYou Quit!\u001b[0m \n')
        else:
            if user_choice == 'R':
                num_user_R += 1
            if user_choice == 'P':
                num_user_P += 1
            if user_choice == 'S':
                num_user_S += 1
            print('Computer play: {} & You play: {}'.format(computer_choice,user_choice))
            message, msg_result, n_games_user_win = who_wins(computer_choice,user_choice, n_games_user_win)
            print('\033[4m\0Round result\033[0m: {} \u001b[1m{}\u001b[0m \n'.format(message,msg_result))

    print('\n\u001b[1mGames played:\u001b[0m {} \n\u001b[1mYou won:\u001b[0m {}\n'.format(n_games,n_games_user_win))
```

## 2 Voters in Florida
---
The file FloridaVoters.html contains a Web Table of republican and democratic voters in various counties in Florida

```python
# Import module
import re
```

```python
def open_file(dir,file_name):
    '''Open a file from a path
        dir : path in which is the file to open
        file_name: name of the file with extension
    '''
    file = open('{}{}'.format(dir,file_name))
    return file
```

```python
def convert_file_to_lines(file):
    '''Create a list of lines of text
        file : archive in format that contain information in text
    '''
    lst_lines = []
    for line in file:
        lst_lines.append(line)
    return lst_lines
```

```python
# Function to get the sorted list of voters
def voters_florida(dir,file):
    '''This function read a html file, get a table of voters in Florida and sort it by number of democrat voters. 
    
        dir :  path of the file
        file : name of the file
    '''
    # Using my open_file function
    FloridaVoters = open_file(dir,file)

    # Step 2: Extract lines in a list

    #Using the function: convert_file_to_lines
    lst_lines = convert_file_to_lines(FloridaVoters)

    # Step 3: Extract info target with re module
    lst_target = []
    for string in lst_lines:
        matches = re.findall('<td>[\w\d\s,-.]+</td>\n', string)
        if len(matches)>0:
            matches = matches[0].lstrip('<td>').rstrip('</td>\n').replace(',','')
            lst_target.append(matches)

    # Step 4: Creating the list of tuples with county name, republican voters, and democrat voters
    lst_tuples = []
    for i,w in enumerate(lst_target):
        if i%6 == 0:
            aux_lst = (lst_target[i],int(lst_target[i+1]),int(lst_target[i+2]))
            lst_tuples.append(aux_lst)

    # Step 5: Creating an auxiliar function to sort
    def getDemocrats(l):
        return l[2]

    # Step 6. Appling the sorted function
    new_sorted_list = sorted(lst_tuples, key=getDemocrats)

    return new_sorted_list
```

```python
def moving_window(string):
    '''Convert a string into 3-word phrases in moving window way
    
        string: chain of words to convert into 3-word phrase
    '''
    lst_phrases = []
    line_words = string.split()
    for i in range(len(line_words)-2):
        phrase = line_words[i] + ' ' + line_words[i+1] + ' ' + line_words[i+2]
        lst_phrases += [phrase]
    return lst_phrases
```

```python
# Extract all the 3-word phrases per each tweet
keys = []
for i,w in enumerate(lst_lines):
    lst_phrases = moving_window(lst_lines[i])
    keys += lst_phrases
keys
#len(keys)
```

```python
# Function to get a dictionary from a list of 3-word phrases
def get_3word_phrase_dict(keys):
    '''Create a dictionary for the unique 3-word phrases.
        keys : list of 3-word phrases
    '''
    # Unique keys
    unique_keys = []
    for i in keys:
        if i not in unique_keys:
            unique_keys.append(i)
            
    # Creating dictionary
    dict_3word_phrase = {}
    for key in unique_keys:
        dict_3word_phrase[key] = 1
        
    return dict_3word_phrase
```

### (b) Calculate similarity between two tweets

```python
# Importing module
import math
```

```python
# Function to get the dictionary of a tweet
def get_dict_tweet(tweet,dict_baseline):
    '''Generate as output a dictinonary with the number of repetitions a tweet has
        tweet :        chain of strings
        dict_baseline: dictionary base in which the tweet is evaluated
    '''
    keys_dict = list(dict_baseline.keys())

    dict_tweet = {}
    for key in keys_dict:
        rep = tweet.count(key)
        dict_tweet[key] = rep
    return dict_tweet
```

```python
# Creating function to calculate cosine similarity:
def cosine(dict_tweet_1,dict_tweet_2):
    '''Calculate the cosine similarity of two dictionaries with the same structure and keys
        dict_tweet_1 : 1st dictionary to compare
        dict_tweet_2 : 2nd dictionary to compare
    '''
    keys_dict = list(dict_tweet_1.keys())
    n_matches = 0
    for key in keys_dict:
        if (dict_tweet_1[key] > 0) and (dict_tweet_2[key] > 0):
            n_matches += 1
    n1 = math.floor(sum([x for x in list(dict_tweet_1.values())]))
    n2 = math.floor(sum([x for x in list(dict_tweet_2.values())]))
    cos = n_matches/(math.sqrt(n1*n2))
    return cos
```

### (c) Read in tweets, and output near-duplicates

```python
def get_near_duplicates(dir,file):
    '''This program allows to find tweets that are near duplicates with the cosine similarity criterion.
        dir:  path in which is the file of tweets
        file: file of tweets
    '''
    # Loading the file
    File_tweets = open_file(dir,file)
    
    # Converting the file into lines
    tweets = convert_file_to_lines(File_tweets)

    # Using moving_window function
    keys = []
    for i,w in enumerate(tweets):
        lst_phrases = moving_window(tweets[i])
        keys += lst_phrases

    # Unique keys
    unique_keys = []
    for i in keys:
        if i not in unique_keys:
            unique_keys.append(i)

    # Creating directory baseline
    dict_3word_phrase = {}
    for key in unique_keys:
        dict_3word_phrase[key] = 1

    for i in range(len(tweets)):
        for j in range(len(tweets)):
            if i != j and i > j:
                tweet_1 = tweets[i]
                dict_tweet_1 = get_dict_tweet(tweet_1,dict_3word_phrase)
                tweet_2 = tweets[j]
                dict_tweet_2 = get_dict_tweet(tweet_2,dict_3word_phrase)
                # Calculate cosine similariry
                cos = cosine(dict_tweet_1,dict_tweet_2)
                cos = int(cos*10000)/100
                if cos > 0.5:
                    print('\u001b[1mTweet (a):\u001b[0m {}, and'.format(tweet_1))
                    print('\u001b[1mTweet (b):\u001b[0m {}'.format(tweet_2))
                    print('Are near-duplicates, with cosine: {}%\n\n'.format(cos))
```

### (a) Build a list of full quotes

```python
def get_full_quotes(dir,file):
    '''Convert file into a list of full quotes
        
        dir :  path or directory of the file
        file : name of the file with extension
    '''
    # Loading file
    file_quotes = open_file(dir,file)
    
    # Converting file to lines
    lst_quotes = convert_file_to_lines(file_quotes)
    
    # Creating list of full quotes : ("quote - speaker")
    full_quotes_list = []
    for i,w in enumerate(lst_quotes):
        if i%2 == 0:
            full_quote = lst_quotes[i].replace('\n','') + ' - ' + lst_quotes[i+1].replace('\n','')
            full_quotes_list.append(full_quote)
    
    return full_quotes_list
```

### (c) Build the postings-list dictionary

```python
def get_unique_words_count(words_lst):
    '''Create a list of tuples with unique words and counts from a list of words.
        
        words_lst : list of words
    '''
    words_lst_unique = []
    lst_aux = []
    for w in words_lst:
        if w not in lst_aux:
            lst_aux.append(w)
            n = words_lst.count(w)
            words_lst_unique += [(w,n)]

    return words_lst_unique
```

```python
def posting_list_dict(full_quotes_list):
    '''Create the complete posting list dictionary from a list of full quotes.
        
        full_quotes_list : complete list of full quotes
    '''
    # Creating empty posting list dict
    dict_posting_list = {}
    for i,w in enumerate(full_quotes_list):
        # full quoute ("quote - speaker")
        key = full_quotes_list[i]
        # getting list of words
        words_lst_aux = full_quote_into_words(key)
        # getting unique list of words and counts
        unique_words_count = get_unique_words_count(words_lst_aux)
        
        # setting auxiliar empty dictionary for each quote
        dict_aux = {}
        # creating the dictionary of the quote
        for k,v in unique_words_count:
            dict_aux[k] = v
        
        # Adding one dictionary for each full quote
        dict_posting_list[key] = dict_aux
        
    return dict_posting_list
```

### (d) Build the reverse postings-list dictionary

```python
def get_big_lst_unique_words(dir,file):
    '''Get the large a complete list of unique words of a file.
    
        dir :  path or directory of the file
        file : name of the file with extension
    '''
    full_quotes_list = get_full_quotes(dir,file)

    big_unique_words_lst = []
    for full_quote in full_quotes_list:
        words_lst = full_quote_into_words(full_quote)
        for w in words_lst:
            if w not in big_unique_words_lst:
                big_unique_words_lst.append(w)

    return big_unique_words_lst
```

```python
def inverse_posting_list(big_unique_words_lst,full_quotes_list):
    '''Create the inverse posting list
        big_unique_words_lst : list of the unique words of the complete file
        full_quotes_list :     list of full quotes
    '''
    dict_inverse_posting_list = {}

    for w in big_unique_words_lst:
        dict_aux = {}
        for full_quote in full_quotes_list:
            # getting list of words
            words_lst_aux = full_quote_into_words(full_quote)
            # getting unique list of words and counts
            unique_words_count = get_unique_words_count(words_lst_aux)
            if w in words_lst_aux:
                dict_aux[full_quote] = [v for k, v in unique_words_count if k == w][0]

        dict_inverse_posting_list[w] = dict_aux
        
    return dict_inverse_posting_list
```

### (e) Write a TF-IDF function

```python
import math
```

```python
def get_tf_idf(word,quote):
    '''Get the TF-IDF 
        word :  word to evaluate
        quote : quote to evaluate
    '''
    if word in list(dict_posting_list[quote].keys()):
        # We calculate TDF getting how many times the word appears in the quote over the max word frequency in the quote
        TF = dict_posting_list[quote][word]/max(dict_posting_list[quote].values())

        # We calculate IDF with log natural of the quotient between number of total quotes 
        # over number of quotes in which the word appears
        IDF = math.log(len(dict_posting_list)/len(dict_inverse_posting_list[word]))

        TF_IDF = TF*IDF
    else:
        TF_IDF = 0
        
    return TF_IDF
```

```python
def get_search_single_word(word):
    '''Make a search of the word and give the tf-idf value for each quote.
        word : word of the list to evaluate
    '''
    dict_single_word = {}
    key_list = list(dict_inverse_posting_list[word].keys())
    for key in key_list:
        dict_single_word[key] = get_tf_idf(word,key)
    return dict_single_word
```

### (g) Quote search using multiple words

```python
def get_search_multiple_words(lst_words):
    '''Make a search of a list of words and give the sum of tf-idf value for each quote.
        lst_words : list of words to evaluate
    '''
    # Getting complete list of quotes
    key_list_aux = []
    for w in lst_words:
        key_list_aux += list(dict_inverse_posting_list[w].keys())
    # Getting unique list of quotes
    key_list = []
    for w in key_list_aux:
        if w not in key_list:
            key_list += [w]
        
    dict_multiple_words = {}
    for key in key_list:
        dict_multiple_words[key] = sum([get_tf_idf(w,key) for w in lst_words])
        
    return dict_multiple_words
```
