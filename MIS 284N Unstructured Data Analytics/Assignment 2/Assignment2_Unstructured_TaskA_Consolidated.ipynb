Top 250

```python
import pandas as pd

wd = webdriver.Chrome('chromedriver',options=chrome_options)
wd.get('https://www.beeradvocate.com/beer/top-rated/')

hrefs=wd.find_elements(By.TAG_NAME, "a")
substring = "/beer/profile"
my_set = set() #### TB - ADDING ELEMENTS TO A SET REMOVES DUPLICATES - ie removing duplicate usernames - duplicates are  not necessary in this specific case 
for lnk in hrefs:
   # get_attribute() to get all href
    full_link = lnk.get_attribute('href')
    if full_link:
      if substring in full_link:
        if len(full_link)>50:
          my_set.add(full_link)

```

```python
comments = pd.DataFrame(columns = ['product_name', 'product_review', 'user_rating']) 
my_set = list(my_set)
```

```python
for item in my_set:
  wd.get(item)
  x = wd.find_element(by=By.CLASS_NAME, value = 'titleBar')
  prod_name = x.find_element(by=By.TAG_NAME, value = 'h1').text
  rev = wd.find_elements(by=By.XPATH, value ='//*[@id="rating_fullview_content_2"]/div')
  rating = wd.find_elements(by=By.XPATH, value ="//*[@id='rating_fullview_content_2']/span[2][contains(@class,'BAscore_norm')]")

  for idx,j in enumerate(rev):
    review = rev[idx].text
    ratings = rating[idx].text
    comments.loc[len(comments.index)] = [prod_name,review, ratings]

#data = comments.drop_duplicates()
```

```python
def get_prod_name(s):
  return s.split("\n")[0]

k = comments['product_name'].map(get_prod_name)
comments['product_name_modified'] = k
```

```python
final_data = comments[['product_name_modified','product_review','user_rating']]
final_data.columns = ['product_name','product_review','user_rating']
```

```python
from google.colab import files
final_data.to_csv('assignment2_final_data.csv')
files.download('assignment2_final_data.csv')
```
