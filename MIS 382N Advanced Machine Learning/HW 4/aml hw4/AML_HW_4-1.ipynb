```python
# Decision Tree Classifier
from sklearn import tree
from sklearn.metrics import confusion_matrix

# Answer here
clf = tree.DecisionTreeClassifier()
clf.fit(X_train, y_train)

y_val_pred = clf.predict(X_val)
y_test_pred = clf.predict(X_test)

print('Confusion matrix of validation set')
print(confusion_matrix(y_val, y_val_pred))

print('Confusion matrix of test set')
print(confusion_matrix(y_test, y_test_pred))
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
# Answer here
logit = LogisticRegression()
logit.fit(X_train, y_train)

y_val_pred = logit.predict(X_val)
y_test_pred = logit.predict(X_test)

print('Confusion matrix of validation set')
print(confusion_matrix(y_val, y_val_pred))

print('Confusion matrix of test set')
print(confusion_matrix(y_test, y_test_pred))
```

```python
# MLP Classifier
from sklearn.neural_network import MLPClassifier
# Answer here
mlp = MLPClassifier(random_state = 15)
mlp.fit(X_train, y_train)

y_val_pred = mlp.predict(X_val)
y_test_pred = mlp.predict(X_test)

print('Confusion matrix of validation set')
print(confusion_matrix(y_val, y_val_pred))

print('Confusion matrix of test set')
print(confusion_matrix(y_test, y_test_pred))

print('Validation set F1 score:', round(metrics.f1_score(y_val, y_val_pred),2))
print('Test set F1 score:', round(metrics.f1_score(y_test, y_test_pred),2))
```

```python
# Hyperparameter tuning
from sklearn.model_selection import PredefinedSplit
parameter_space = {
    'hidden_layer_sizes': [(20,20,20), (20,20), (20,)],
    'activation': ['tanh', 'relu','logistic'],
    'solver': ['sgd', 'adam'],
    'alpha': [0.0001, 0.0005, 0.001, 0.005],
    'learning_rate': ['constant','adaptive'],
    'random_state' : [20]
}

# Create a list where train data indices are -1 and validation data indices are 0
split_index = [-1 if x in X_train.index else 0 for x in X_1.index]
ps = PredefinedSplit(split_index)

from sklearn.model_selection import GridSearchCV
# Let's train the grid search models on X_train but measure it's effectiveness on X_val. We can do this using the 
# PredefinedSplit function
gr = GridSearchCV(mlp, parameter_space, n_jobs=-1, cv=ps, scoring='f1') 
# We're using F1 score as it is a better metric for measuring effectiveness of models on unbalanced datasets
gr.fit(X_1, y_1)
```

```python
print('Best parameters:\n', gr.best_params_)
```

```python
mlp_best = gr.best_estimator_
mlp_best.fit(X_train, y_train)

y_val_pred = mlp_best.predict(X_val)
y_test_pred = mlp_best.predict(X_test)

print('Confusion matrix of validation set')
print(confusion_matrix(y_val, y_val_pred))

print('Confusion matrix of test set')
print(confusion_matrix(y_test, y_test_pred))

print('Validation set F1 score:', round(metrics.f1_score(y_val, y_val_pred),2))
print('Test set F1 score:', round(metrics.f1_score(y_test, y_test_pred),2))
```
